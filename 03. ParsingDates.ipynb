# Parsing de Fechas

En este JNB se revisará el tema de Parsing de Fechas.

```python
# modules we'll use
import pandas as pd
import numpy as np
import seaborn as sns
import datetime

# read in our data
landslides = pd.read_csv("https://raw.githubusercontent.com/GTD2101-ACD-AA-EJ23/04-pandas-clean-ricardo-qm/main/data/catalog.csv")

# set seed for reproducibility
np.random.seed(0)
```

# Verficando el tipo de dato de la columna Fecha.

Validar que efectivamente las columnas de fechas almacenan datos de este tipo.

```python
# print the first few rows of the date column
print(landslides['date'].head())
```

```python
landslides['date'].dtype
```

# Convertir la columna al tipo datetime

```python
# create a new column, date_parsed, with the parsed dates
landslides['date_parsed'] = pd.to_datetime(landslides['date'], format="%m/%d/%y")
```

```python
landslides
```

```python
landslides['date_parsed'].head()
```

# Seleccionando el día de la fecha

Ya teniendo la columna con tipo fecha, podemos acceder a los elementos de la misma.

```python
# get the day of the month from the date_parsed column
day_of_month_landslides = landslides['date_parsed'].dt.day
day_of_month_landslides.head()
```

```python
# remove na's
day_of_month_landslides = day_of_month_landslides.dropna()

# plot the day of the month
sns.distplot(day_of_month_landslides, kde=False, bins=31)
```
